```python
#ToDo class
class Task:
    def __init__(self, title, description, duedate, status='Incomplete'):
        self.title = title 
        self.description = description 
        self.duedate = duedate 
        self.status = status
    def __str__(self):
        return f'Title: {self.title}\n Description: {self.description}\n Due date: {self.duedate}\n Status: {self.status}'
    

class TodoList:
    def __init__(self):
        self.tasks = []

    def newtask(self, ntask):
        self.tasks.append(ntask) 
        print(f'Task {ntask.title} has been added to the list')

    def status(self, title):
        for x in self.tasks:
            if x.title.lower() == title.lower():
                x.status = "Completed"
                print(f'The task {title} marked as Completed')
                return 
        print('Task {title} is not found')
    
    def alltasks(self):
        if self.tasks == []:
            print("There is no task in the list")
            return
        print("Here are all tasks: ")
        for index, ntask in enumerate(self.tasks, 1):
            print(f'Task {index}:')
            print(ntask)

    def incomplete(self):
        itasks = [task for task in self.tasks if task.status == "Incomplete"]
        if itasks == []:
            print("There is no Incomplete task")
            return 
        print("Incomplete tasks: ")
        for index, ntask in enumerate(itasks, 1):
            print(f"Task {index}:\n {ntask}")
    
from IPython.display import clear_output
import time 

def todo():
    print("Welcome to To-do-list App menu: ")
    print("1. Add new task to the list")
    print("2. List all tasks")
    print("3. Mark task as completed")
    print("4. List only incomplete tasks")
    print("5. Exit") 
    
def main():
    todolist = TodoList()

    while True:
        todo()
        choice = input("Enter a number between 1 and 5: ")

        if choice == '1':
            title = input("Enter a title of the task: ")
            description = input('Enter a description of the task: ')
            duedate = input("Enter due date(format:YYY-MM-DD): ")
            newtask = Task(title,description, duedate)
            todolist.newtask(newtask) 
        elif choice == '2':
            todolist.alltasks()
        elif choice == '3':
            utitle = input("Enter title to be marked as Completed: ")
            todolist.status(utitle)
        elif choice == '4':
            todolist.incomplete()
        elif choice == '5':
            print("Exiting the program")
            break 
        else: 
            print("Invalid choice! Please enter number between 1 and 5")
        clear_output(wait=True)
        time.sleep(3)
main()
```
